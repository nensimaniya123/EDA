```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("Titanic-Dataset.csv") 
df.head()
```

```python
# Histogram
df[['Age', 'Fare']].hist(bins=20, figsize=(10, 4), color='skyblue')
plt.suptitle('Histograms of Age and Fare')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('Boxplots of Age and Fare')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.suptitle('Pairplot of Key Features', y=1.02)
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Count by Class")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['Survived']==1], x='Age', label='Survived', shade=True)
sns.kdeplot(data=df[df['Survived']==0], x='Age', label='Not Survived', shade=True)
plt.title("Age Distribution by Survival")
plt.legend()
plt.show()
```
